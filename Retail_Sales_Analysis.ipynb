## **Feature Engineering**

Feature engineering is performed to create additional variables from existing data. These newly generated features help uncover time-based patterns and support deeper business analysis.

In this section, date-related variables such as Month, Day, and Quarter are extracted from the transaction date to enable trend and seasonal analysis.

```python
# ================================================
# Feature Engineering
# ================================================

# Extract 'Month' from Order_Date
df['Month'] = df['Order_Date'].dt.month_name()

# Extract Day Name
df['Day_Name'] = df['Order_Date'].dt.day_name()

# Extract Quarter
df['Quarter'] = df['Order_Date'].dt.quarter

# Display newly created columns
df[['Order_Date','Month', 'Day_Name', 'Quarter']].head()
```

**Observations:**

- Three new time-based variables were created from the **Order_Date** column:
    - Month
    - Day_Name
    - Quarter

- The **Month** variable enables monthly sales trend analysis and seasonal pattern identification.

- The **Day_Name** variable allows weekday-level analysis and helps identify customer purchasing behavior across different days.

- The **Quarter** variable supports broader performance comparisons across business periods.

- Feature engineering transforms raw date information into meaningful business variables that improve analytical capabilities.

Overall, the newly generated variables create opportunities for time-series and trend-based analysis.

## Business Question 1

### Which region generated the highest total sales?

This analysis evaluates total sales generated across different regions to identify the strongest revenue-contributing market.

```python
## Sales by region

region_sales = df.groupby('Region')['Final_Sales'].sum().sort_values(ascending=False)
```

### Visualization

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=region_sales.index,
    y=region_sales.values
)

plt.title('Total Sales by Region')

plt.xlabel('Region')

plt.ylabel('Total Sales')

for i, value in enumerate(region_sales.values):
    plt.text(i, value + 5000, f'{value:,.0f}', ha='center')

plt.show()
```

## Observations

After analyzing total sales across regions, the following observations were identified:

- The **West region** generated the highest total sales, contributing approximately **₹805,000** in revenue.

- The **South region** recorded the second-highest sales performance, generating approximately **₹782,000**.

- The **East region** contributed moderate sales, with total revenue of approximately **₹650,000**.

- The **North region** generated the lowest sales among all regions, contributing approximately **₹568,000**.

- The difference between the highest-performing region (West) and the lowest-performing region (North) is significant, indicating regional variations in customer demand and sales performance.

- West and South regions appear to be key revenue-generating markets and may represent important growth opportunities for the business.

Overall, regional sales performance is not evenly distributed, suggesting that management should investigate factors contributing to the stronger performance of West and South regions.

## Business Question 3

### What is the Relationship Between Discount Percentage and Profit?

This analysis investigates whether discounts influence profitability and examines the relationship between promotional strategies and profit generation.

```python
# Relationship between Discount and Profit

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='Discount_Percent',
    y='Profit'
)

plt.title('Discount Percentage vs Profit')
plt.xlabel('Diacount Percentege')
plt.ylabel('Profit')

plt.show()
```

```python
# Correlation between Discount and Profit

correlation = df['Discount_Percent'].corr(df['Profit'])

print("Correlation:", round(correlation,3))
```

### Interpretation Guide

- 0.70 to 1.00   Strong Positive
- 0.30 to 0.69   Moderate Positive
- 0.00 to 0.29   Weak Positive

- 0             No Relationship

- -0.29 to 0.00 Weak Negative
- -0.69 to -0.30 Moderate Negative
- -1.00 to -0.70 Strong Negative

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=df,
    x='Discount_Percent',
    y='Profit'
)

plt.title('Discount Percentage vs Profit')
plt.xlabel('Diacount Percentege')
plt.ylabel('Profit')

plt.show()
```

### Observations

- The correlation between Discount Percentage and Profit is **-0.139**, indicating a weak negative relationship.

- As discount percentages increase, profit tends to decrease slightly; however, the relationship is not strong.

- Profitability varies considerably across all discount levels, suggesting that other factors also influence profit.

- High-profit transactions are observed even at higher discount levels, indicating that discounts do not always reduce profitability.

- Factors such as product category, quantity sold, pricing strategy, and customer demand may have a greater impact on profit than discounts alone.

Overall, discount percentage has only a minor influence on profitability within this dataset.

### Key Business Insight

While discounts are generally expected to reduce profitability, the analysis shows only a weak negative relationship between discounts and profit. This suggests that the company can still generate profitable transactions even when discounts are offered, provided that pricing and product strategies are managed effectively.
